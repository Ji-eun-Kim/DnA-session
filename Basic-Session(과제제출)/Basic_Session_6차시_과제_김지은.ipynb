### Q1. 내장을 제외한 전복의 무게를 구하는 함수를 정의하고, 
### 함수를 적용해서 '전복-내장'이라는 컬럼을 abalone에 추가하세요 (11점)

```python
# 함수를 정의하세요 (6점)
# whole_weight - viscera_weight

def abalone_weight(x,y):
    return x-y
```

```python
# 데이터에 함수를 적용하고, '전복-내장' 컬럼을 추가하세요 (5점)
abalone_viscera= abalone_weight(abalone['whole_weight'], abalone['viscera_weight'])

abalone['abalone_viscera']= abalone_viscera
```

### Q2 . apply에서 lambda를 이용해서 'whole_weight'의 평균을 넘는 지에 대한 새로운 컬럼('heavy_w')를 정의하고, groupby를 이용하여 'sex'별 'heavy_w'가 평균을 넘는 개수가 몇개인지 구하세요 (11점)

```python
# apply에서 lambda를 이용해서 'whole_weight'의 평균을 넘는 지에 대한 새로운 컬럼('heavy_w')를 정의하세요 (6점)

abalone['heavy_w']= abalone.groupby('sex')['whole_weight'].apply((lambda x: x.mean()< x));abalone['heavy_w']
```

```python
# groupby를 이용하여 'sex'별 'heavy_w'가 평균을 넘는 개수가 몇개인지 구하세요 (5점)
qur= 'heavy_w ==True'
abalone.query(qur).groupby('sex')['heavy_w'].count()
```

### Q5. sex를 제외한 column 중 원하는 두 column을 정하고, 두 컬럼의 합을 구하는 groupby를 만들기  (12점)
집계함수, apply, agg, lambda 등 사용하여 두개를 작성하세요  
(컬럼과 함수는 다른것으로 사용하세요)  
(key값은 sex로)

```python
# (1) (6점)
# 선택 컬럼을 적어주세요 :  whole_weight, shucked_weight

# 코드를 작성하세요: 


abalone.groupby('sex')[['whole_weight','shucked_weight']].apply(lambda x: x.sum())
```

```python
# (2) (6점)
# 선택 컬럼을 적어주세요 :  height, whole,weight

# 코드를 작성하세요

abalone.groupby('sex')[['height','whole_weight']].agg(sum)
```

### Q6. key값을 sex로 하고, 나머지 column들에 min, mean, max를 적용하고, 이름을 최소, 평균, 최대 로 바꾸기 (8점)

```python
# 코드를 작성하세요

abalone.groupby('sex')['sex','length','diameter','height','whole_weight','shucked_weight','viscera_weight','shell_weight','rings','abalone_viscera'].agg([('최소','min'), ('평균','mean'),('최대','max')])
```

### Q7. pivot_table을 이용하여 abalone 데이터 살펴보기  (10점)
pivot_table을 사용하여 두가지를 작성해보세요

#### 9-1. pivot_table을 이용하여 작성하세요 (6점)

```python
# pivot_table 사용한 두가지 방법

pd.pivot_table(df_phone,
              values= 'duration' ,
               index= ['month','item'],
              columns= 'network',
               aggfunc=sum,
              fill_value=0)
```

```python
df_phone.pivot_table(values= 'duration',
               index= ['month','item'],
              columns= 'network',
               aggfunc=sum,
              fill_value=0)
```

#### 9-2. crosstab을 이용하여 작성하세요 (6점)
(crosstab의 경우 뒤에 소수점이 있을 수 있습니다)  

```python
# crosstab 사용한 코드를 작성해주세요

pd.crosstab(index=[df_phone['month'],df_phone['item']],
            columns=df_phone['network'],
            values=df_phone['duration'],
            aggfunc='sum'
           )
```
